```python
import cv2
import mediapipe as mp
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_pose = mp.solutions.pose
```

```python
IMAGE_FILES = [path+'/'+os.listdir(path)[0]]
BG_COLOR = (192, 192, 192)

# Initialize MediaPipe Pose with desired configurations
with mp_pose.Pose(
        static_image_mode=True,
        model_complexity=2,
        enable_segmentation=True,
        min_detection_confidence=0.5) as pose:

    for idx, file in enumerate(IMAGE_FILES):
        # Read the image
        image = cv2.imread(file)
        image_height, image_width, _ = image.shape

        # Process the image with MediaPipe Pose
        results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

        # Check if pose landmarks are detected
        if not results.pose_landmarks:
            continue

        # Print the coordinates of the nose landmark
        print(
            f'Nose coordinates: ('
            f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].x * image_width}, '
            f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].y * image_height})'
        )

        # Create an annotated version of the image
        annotated_image = image.copy()

        # Segment the person from the background
        condition = np.stack((results.segmentation_mask,) * 3, axis=-1) > 0.1
        bg_image = np.zeros(image.shape, dtype=np.uint8)
        bg_image[:] = BG_COLOR
        annotated_image = np.where(condition, annotated_image, bg_image)

        # Draw pose landmarks on the annotated image
        mp_drawing.draw_landmarks(
            annotated_image,
            results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style())

        # Save the annotated image
        cv2.imwrite('/tmp/' + str(idx) + '.png', annotated_image)

        # Plot the pose world landmarks (optional)
        mp_drawing.plot_landmarks(
            results.pose_world_landmarks, mp_pose.POSE_CONNECTIONS)
```

```python
import cv2
import mediapipe as mp
from google.colab.patches import cv2_imshow
import pandas as pd
import numpy as np

# Initialize MediaPipe Pose
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# 좌표 데이터프레임 생성
coordinate_df = pd.DataFrame()

# 이미지 파일의 경로
for i in imgs:
  IMAGE_FILES = [path+'/'+i]


  # MediaPipe Pose 초기화
  with mp_pose.Pose(
          static_image_mode=True,
          model_complexity=2,
          min_detection_confidence=0.5) as pose:

      for idx, file in enumerate(IMAGE_FILES):
          img_df = pd.DataFrame({'img_num': [i]})
          # 이미지 읽기
          image = cv2.imread(file)
          image_height, image_width, _ = image.shape

          # MediaPipe Pose로 이미지 처리
          results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

          # 랜드마크가 감지되었는지 확인
          if results.pose_landmarks:
            # 랜드마크를 이미지 위에 그리기
              mp_drawing.draw_landmarks(
                image,
                results.pose_landmarks,
                mp_pose.POSE_CONNECTIONS)

            # 랜드마크의 좌표 출력
              print("Landmark coordinates for", file)
              for landmark_id, landmark in enumerate(results.pose_landmarks.landmark):
                  landmark_x = int(landmark.x * image_width)
                  landmark_y = int(landmark.y * image_height)
                  landmark_z = landmark.z

                  # 랜드마크 레이블 가져오기
                  landmark_label = mp_pose.PoseLandmark(landmark_id).name

 #               print(f"Landmark {landmark_label} - X: {landmark_x}, Y: {landmark_y}, Z: {landmark_z}")
                  array_coordinate = np.array([landmark_x, landmark_y, landmark_z])
                  img_df.at[0, landmark_id] = array_coordinate.astype(object)

              coordinate_df = pd.concat([coordinate_df, img_df])
            # 이미지를 화면에 표시
#              cv2_imshow(image)

          else:
              print("No pose landmarks detected in", file)
```

```python
coordinate_df.reset_index(drop=True, inplace=True)
```
