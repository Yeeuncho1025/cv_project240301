```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


path = '/content/drive/MyDrive/CV_img_data/'
df_true = pd.read_csv(path+'keypoint_data_t_angle.csv', index_col=0)
df_false = pd.read_csv(path+'keypoint_data_f_angle.csv', index_col=0)
need_col = ['img_num', 'right_angle[12]',
       'right_angle[14]', 'right_angle[16]', 'right_angle[18]',
       'right_angle[20]', 'right_angle[22]', 'right_angle[24]',
       'right_angle[26]', 'right_angle[28]', 'left_angle[11]',
       'left_angle[13]', 'left_angle[15]', 'left_angle[17]', 'left_angle[19]',
       'left_angle[21]', 'left_angle[23]', 'left_angle[25]', 'left_angle[27]']

df_true = df_true[need_col]
df_false = df_false[need_col]

df_true.at[:,'anno']= 1
df_false.at[:,'anno']= 0
df = pd.concat([df_true, df_false])
df.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

df_data = df.drop(['img_num'], axis=1)
df_data = df_data.iloc[:,[7,8,16,17,18]]
data = df_data.values
X = data[:,:-1]
y = data[:,-1]

train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from xgboost import XGBClassifier

skl_xgb = XGBClassifier(learning_rate=0.000955,n_estimators=100,max_depth=3)
skl_xgb.fit(train_x, train_y)

y_train_pred = skl_xgb.predict(train_x)
y_pred = skl_xgb.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score

def classifier_eval(y_test, y_pred):
  print('accuracy: ', accuracy_score(y_test, y_pred))
  print('recall: ', recall_score(y_test, y_pred))
  print('precision: ', precision_score(y_test, y_pred))
  print('f1: ', f1_score(y_test, y_pred))
  print('roc_auc: ', roc_auc_score(y_test, y_pred))

classifier_eval(train_y, y_train_pred)
print('')
classifier_eval(test_y, y_pred)
```
